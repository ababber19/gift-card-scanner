```python
import numpy as np
import pandas as pd
import os, shutil
```

```python
df = pd.read_csv("../input/hotstar-gift-cards/Costco Cards.csv")
```

```python
# Gets names of images available
base_dir_input = "../input/hotstar-gift-cards/allPicturesCombined_cleaned/allPicturesCombined_cleaned"
file_names = (os.listdir(base_dir_input))
```

```python
# Limits df to only rows with available images, only single cards
df = df[df["Image Name"].isin(file_names)]
df = df.drop_duplicates(subset=["Image Name"])
df = df[df["isSingleCard"]==1]
df = df.drop(columns=["Address", "isSingleCard", "isPacket"])
```

# Run If You Only Have Access to Original Images

```python
 # Creates smaller data set
base_dir_input = "../input/hotstar-gift-cards/allPicturesCombined_cleaned/allPicturesCombined_cleaned"
base_dir_output = "./"

# Creates directory for only singleCards
isSingleCard = os.path.join(base_dir_output, "isSingleCard")

try:     
    os.mkdir(isSingleCard)

except:
    print("Directories already exists")
```

```python
# Creates output directory with images that we can use
base_dir_output = "./isSingleCard"
for name in file_names:
    if name in df["Image Name"].tolist():    
        src = os.path.join(base_dir_input, name) 
        dst = os.path.join(base_dir_output, name) 
        shutil.copyfile(src, dst)
        
len(os.listdir("./isSingleCard")) == len(df["Image Name"])
```

# Run After Data Input Correction

```python
 # Creates smaller data sets
base_dir_input = "../input/hotstar-only-single-cards/isSingleCard"
base_dir_output = "./"

# Creates directories for training, validation, and testing
train_dir = os.path.join(base_dir_output, 'train')
validation_dir = os.path.join(base_dir_output, 'validation')
test_dir = os.path.join(base_dir_output, 'test')

try:
    os.mkdir(train_dir)
    os.mkdir(validation_dir)
    os.mkdir(test_dir)


except:
    print("Directories already exists")
```

```python
# Gets desired length of each data set
train_len = int(round(len(df)*.8,0))
validation_len = int(round(len(df)*.1,0))
testing_len = int(len(df)- train_len - validation_len)
```

```python
# CREATES TRAINING DAA SET
# Gets names of images available
file_names = (os.listdir(base_dir_input))

# Creates training data set
fnames = ['{}'.format(file_names[i]) for i in range(train_len)] 
for fname in fnames:
    src = os.path.join(base_dir_input, fname)
    dst = os.path.join(train_dir, fname) 
    shutil.copyfile(src, dst)
```

```python
# Creates validation data set
fnames = ['{}'.format(file_names[i]) for i in range(train_len, train_len+validation_len)] 
for fname in fnames:
    src = os.path.join(base_dir_input, fname)
    dst = os.path.join(validation_dir, fname) 
    shutil.copyfile(src, dst)
    
    
# Creates testing data set
fnames = ['{}'.format(file_names[i]) for i in range(train_len+validation_len, train_len+validation_len+testing_len)] 
for fname in fnames:
    src = os.path.join(base_dir_input, fname)
    dst = os.path.join(test_dir, fname) 
    shutil.copyfile(src, dst)
```

```python
df.to_csv('cards_labels.csv',index=False)
```
